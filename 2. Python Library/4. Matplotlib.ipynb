## matplotlib
 - 2D 그래프로 시각화를 가능하게 해주는 라이브러리
 - 파이썬에서 데이터를 차트나 플롯으로 그려줌
 - 차트의 종류 : 산점도, 바, 파이, 히스토그램, 꺾은선 그래프 등 다양함
 
### 데이터 시각화란?
- 광범위하게 분산된 방대한 양의 자료를 한 눈에 볼 수 있도록 도표나 차트 등으로 정리하는 것
- 시각화를 통해 데이터의 특징을 쉽게 파악할 수 있다
 

```python
# matplotlib 라이브러리 불러오기
# matplotlib 패키지의 pyplot 모듈을 사용
import matplotlib.pyplot as plt
```

### 기본적인 그래프 그려보기

```python
# Line plot (선 그래프)
y = [2, 12, 6]
plt.plot(y)

# plot (좌표 y요소)
# x좌표는 0부터 1씩 증가시키며 자동 할당

# print()와 같은 역할을 하는 명령어:  show()
# 다 그렸다면 꼭 마무리를 show() 사용해서 출력
plt.show()
```

```python
# x축 y축 설정해서 그려보기
x = [1,2,3]
y = [2,4,6]
# plt.plot(x,y)

# 선의 스타일 입히기
plt.plot(x,y,ls = '-')
plt.show()
```

```python
# x축 y축 설정해서 그려보기
x = [1,2,3]
y = [2,4,6]

# marker 찍기
plt.plot(x,y, ls= '--', marker = 's', lw = 5)
plt.show()
```

```python
# x축 y축 설정해서 그려보기
x = [1,2,3]
y = [2,4,6]

# 선의 색상 선택
plt.plot(x,y, ls= '--', marker = '*', lw = 1, color = 'r')
plt.show()
```

```python
# x축 y축 설정해서 그려보기
x = [1,2,3]
y = [2,4,6]

# 마커의 색상 선택

# 속성값
# 1. 라인의 스타일 (Line Style) 설정 : ls = '문자열'
# 2. 라인의 두깨 (Line Width) 설정 : lw = int
# 3. 선의 색깔(Color) 바꾸기 : c = 'r,g,b'
# 4. 값의 위치 표시 : marker = '문자열'
# 5. 마커 내부 색깔 (Marker Face Color) : mfc = '문자열'
# 6. 마커 선 색깔 (Marker Edge Color) : mec = '문자열'
# 7. 마커 크기 (Marker Size) : ms = int
# 8. 마커 선 굵기 (Marker Edge Width) : mew = int
plt.plot(x,y, ls= '--', marker = '*', lw = 1, color = 'r', ms = 8, mec = 'g', mfc = 'blue')
plt.show()
```

### 선 그래프 실습문제
- x = np.arange(8)
- y = [5,2,3,6,8,7,5,7]

```python
import numpy as np
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,c='m' , ls = '-.', lw = 4, marker = '$❤$', ms = 10, mec = 'c', mfc = 'y', mew = 2)


# 그래프의 숫자표시 범위 지정 (xlim, ylim)
# 범위를 지정해서 그래프 확대/축소해서 출력 가능
plt.xlim(-2,10)
plt.ylim(-5,12)
plt.show()
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]
plt.plot(x,y,c='m' , ls = '-.', lw = 4, marker = '$❤$', ms = 10, mec = 'c', mfc = 'y', mew = 2)


# 그리드 그리는 방법 (그리드 : 모눈종이, 격자눈금) --> 그래프를 보기 쉽게 만들어줌
plt.grid()
plt.yticks(y)
plt.xticks(x)
plt.show()
```

### 여러개의 차트를 하나의 공간에 그리기 

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')

# 범례 설정하기
# 범례 : (참고사항) 데이터의 종류를 표시하기 위한 텍스트
plt.legend()

# 차트 제목 설정하기
plt.title('chart', loc = 'center')

# x축, y축 이름
plt.xlabel('x', {'fontname': 'Times New Roman',
     'fontsize': 30,
     'style': 'italic', # 'oblique'
      'fontweight': 'bold'})
plt.ylabel('y')


plt.show()
```

### 공공데이터 분석 및 시각화
- 2017년 전국 사망교통사고 데이터

```python
# 데이터 분석에 필요한 라이브러리 로딩
# numpy, pandas, matplotlib 불러오기 , 한글 인코딩
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
# 실제 데이터 로딩
data = pd.read_csv('Traffic_Accident_2017.csv', encoding = 'euc-kr')
data
```

```python
# 한글적용
rc('font', family = 'Malgun Gothic')
```

```python
# 데이터 규모 확인
data.info()
```

```python
# 필요한 데이터만 추출
# .value_counts() : 하나의 컬럼 안에 있는 값들을 중복 제거 후 세어주는 함수
data['사망자수'].value_counts()
```

```python
# 데이터 구조 살피기 (최상단 5개만 출력)
data.head()
# 하단 5줄
data.tail()
```

```python
# 컬럼(열)명만 추출
data.columns
```

### 요일별 사고건수 카운트
- 요일 순서대로 정렬 후 y변수 저장

```python
# 요일별 발생한 사고 건수를 세어서 temp에 담아주기
temp = data['요일'].value_counts()
temp
```

```python
# 유니크 값으로 설정(순서대로 정렬) 후 y변수에 담아주기
# 요일별 정렬 (인덱스 위치 강제)
y = temp[['월', '화', '수','목', '금', '토', '일']]
```

### 시각화
- 바 그래프

```python
# y축 범위 설정 : 500~ 620
# x축 y축의 이름 설정
# title : 전체 그래프 이름 설정
# 요일별 교통사고 건수
x = range(7)
color = ['r','g','b','r','g','b','r']
plt.ylim(500,620)
plt.bar(x,y, color = color)

plt.xlabel('요일')
plt.ylabel('건수', rotation = 0)
plt.title('요일별 교통사고 건수', loc = 'center')

plt.xticks(x, ['월', '화', '수','목', '금', '토', '일'])
plt.yticks(y)

# 그래프의 데이터 수치 표현: plt.text()
for v, i in enumerate(x):
    plt.text(v, y[i], y[i], fontsize = 9, color = 'b', horizontalalignment = 'center', verticalalignment = 'bottom')
    # 좌표가 x축은 v씩 늘어나고 y 축은 i 씩 늘어나게 만듬 / horizontalalignment = 텍스트의 수평위치선정 (left, center, right)
    #                                                       verticalalignment = 텍스트의 수직 위치선청( top, center, bottom)
plt.show()
```
